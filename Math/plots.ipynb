# Matplotlib Plotting Examples

This notebook demonstrates the core plotting capabilities of matplotlib, the most widely used plotting library in Python. Each section introduces a different plot type with a short description and a code example.

## 1. Simple Line Plot

A line plot is the most basic plot type, useful for visualizing functions or trends. Here, we plot a sine wave.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
plt.figure()
plt.plot(x, y, label='sin(x)')
plt.title('Simple Line Plot')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)
plt.show()
```

## 2. Scatter Plot

Scatter plots are useful for visualizing the relationship between two sets of data points.

```python
np.random.seed(0)
x = np.random.rand(50)
y = np.random.rand(50)
plt.figure()
plt.scatter(x, y, c='red', marker='o', label='Random Points')
plt.title('Scatter Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## 3. Bar Plot

Bar plots are used to compare quantities across different categories.

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 18]
plt.figure()
plt.bar(categories, values, color='skyblue')
plt.title('Bar Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

## 4. Histogram

Histograms show the distribution of a dataset by grouping values into bins.

```python
samples = np.random.randn(1000)
plt.figure()
plt.hist(samples, bins=30, color='green', alpha=0.7)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

## 5. Multiple Subplots

You can display several plots in a single figure using subplots. This is useful for comparing different functions or datasets side by side.

```python

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

# Data for plotting
x = np.linspace(0, 2 * np.pi, 100)
signals = [
    (np.sin(x), 'sin(x)', 'blue'),
    (np.cos(x), 'cos(x)', 'red'),
    (np.tan(x), 'tan(x)', 'green'),
    (np.exp(-x), 'exp(-x)', 'black')
]

# Plot each signal in its corresponding subplot
positions = [(0, 0), (0, 1), (1, 0), (1, 1)]
for (y, title, color), (i, j) in zip(signals, positions):
    axs[i, j].plot(x, y, color=color)
    axs[i, j].set_title(title)
    axs[i, j].set_xlabel('x')
    axs[i, j].set_ylabel('y')
    axs[i, j].grid(True)

plt.tight_layout()
plt.show()
```

## 6. Customizing Styles

Matplotlib supports many built-in styles to quickly change the look of your plots. Here we use the 'ggplot' style.

```python
plt.style.use('ggplot')
x = np.linspace(0, 10, 100)
y1 = np.exp(-x / 3) * np.sin(x)
y2 = np.exp(-x / 3) * np.cos(x)
plt.figure()
plt.plot(x, y1, label='Damped Sine')
plt.plot(x, y2, label='Damped Cosine')
plt.title('Custom Style: ggplot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

# Reset to default style for future plots
plt.style.use('default')
```

## 7. Plotting Phasors

Phasors (complex numbers) can be visualized as arrows in the complex plane. This is useful in electrical engineering to represent sinusoidal signals with magnitude and phase. The following example shows how to plot several phasors as arrows from the origin.

```python
# Example: Plotting phasors (complex numbers) as arrows
current = 0.5 + 0.4j
impedance = 2 + 3j
voltage = current * impedance

phasors = [voltage, current, impedance]
colors = ['red', 'blue', 'green']

plt.figure(figsize=(6, 6))
max_radius = max(abs(z) for z in phasors) * 1.2  # 20% margin
for z, color in zip(phasors, colors):
    plt.arrow(0, 0, np.real(z), np.imag(z), head_width=0.1, head_length=0.15, length_includes_head=True, color=color)
    plt.plot(np.real(z), np.imag(z), 'o' , color= 'black')
    plt.text(np.real(z)*1.05, np.imag(z)*1.05, f'{np.abs(z):.2f}∠{np.angle(z, deg=True):.0f}°', fontsize=10, color=color)
plt.axhline(0, color='gray', lw=1)
plt.axvline(0, color='gray', lw=1)
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.title('Phasors in the Complex Plane')
plt.grid(True)
plt.axis('equal')
plt.xlim(-max_radius, max_radius)
plt.ylim(-max_radius, max_radius)
plt.show()
```
